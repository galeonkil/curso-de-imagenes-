```python
# Fijar la semilla para asegurar la reproducibilidad
np.random.seed(123)

# Generar la matriz aleatoria de 16x16 con valores 0 o 1
watermark1 = np.random.randint(0, 2, (16, 16))


plt.imshow(watermark1, cmap='gray')
plt.title('Marca de Agua 2')
plt.show()
```

```python
# Preprocesamiento de la imagen
def preprocess_image(image):
    feature_extractor = ViTFeatureExtractor.from_pretrained('google/vit-base-patch16-224')
    return feature_extractor(images=image, return_tensors="pt").pixel_values

# Función para codificar la marca de agua
def encode_watermark(image, watermark):
    input_image = preprocess_image(image)

    # Convertir marca de agua a un tensor y redimensionar
    watermark_tensor = torch.tensor(watermark, dtype=torch.float32).unsqueeze(0).unsqueeze(0)
    watermark_tensor = torch.nn.functional.interpolate(watermark_tensor, size=(input_image.shape[2], input_image.shape[3]))
    watermark_tensor = watermark_tensor.squeeze().expand_as(input_image[:, 0, :, :])

    # Codificar la marca de agua en el canal de la imagen
    stego_image = input_image.clone()
    stego_image[:, 0, :, :] = input_image[:, 0, :, :] + 0.01 * watermark_tensor

    return stego_image

# Función para decodificar la marca de agua
def decode_watermark(stego_image, original_image):
    # Suponemos que la marca de agua está codificada en el primer canal de la imagen
    extracted_watermark = (stego_image[:, 0, :, :] - original_image[:, 0, :, :]) / 0.01
    return extracted_watermark

# Verificación de la marca de agua con el modelo ViT
def verify_watermark(image, stego_image, model):
    original_features = model(preprocess_image(image)).logits
    stego_features = model(stego_image).logits

    # Comparar las características de la imagen original y la imagen esteganografiada
    diff = torch.abs(original_features - stego_features).mean()
    print(f"Diferencia media: {diff.item()}")  # Añadir impresión de la diferencia media
    return diff.item() < 1e-1  # Ajustar el umbral de diferencia aceptable

# Función para mostrar la comparación de las imágenes y las marcas de agua
def show_comparison(original_image, stego_image, original_watermark, extracted_watermark):
    fig, axs = plt.subplots(2, 2, figsize=(10, 10))

    # Mostrar imagen original
    original_image_np = original_image.squeeze().permute(1, 2, 0).detach().numpy()
    axs[0, 0].imshow(original_image_np)
    axs[0, 0].set_title('Imagen Original')
    axs[0, 0].axis('off')

    # Mostrar imagen esteganografiada
    stego_image_np = stego_image.squeeze().permute(1, 2, 0).detach().numpy()
    axs[0, 1].imshow(stego_image_np)
    axs[0, 1].set_title('Imagen con Marca de Agua')
    axs[0, 1].axis('off')

    # Mostrar marca de agua original
    axs[1, 0].imshow(original_watermark, cmap='gray')
    axs[1, 0].set_title('Marca de Agua Original')
    axs[1, 0].axis('off')

    # Mostrar marca de agua extraída
    extracted_watermark_np = extracted_watermark.squeeze().detach().numpy()
    axs[1, 1].imshow(extracted_watermark_np, cmap='gray')
    axs[1, 1].set_title('Marca de Agua Extraída')
    axs[1, 1].axis('off')

    plt.show()

# Función para guardar la imagen esteganografiada
def save_stego_image(stego_image, filename):
    stego_image_np = stego_image.squeeze().permute(1, 2, 0).detach().numpy()
    stego_image_pil = Image.fromarray((stego_image_np * 255).astype(np.uint8))
    stego_image_pil.save(filename)

model = ViTForImageClassification.from_pretrained('google/vit-base-patch16-224')
image = Image.open('sample_data/robot.jpg')
np.random.seed(123)
watermark = np.random.randint(0, 2, (16, 16))  # Marca de agua binaria aleatoria

# Mostrar marca de agua original
show_watermark(watermark, 'Marca de Agua Original')

original_image_tensor = preprocess_image(image)
stego_image = encode_watermark(image, watermark)
extracted_watermark = decode_watermark(stego_image, original_image_tensor)

# Mostrar comparación de las imágenes y las marcas de agua
show_comparison(original_image_tensor, stego_image, watermark, extracted_watermark)

# Guardar stego imagen
save_stego_image(stego_image, 'stego_image.jpg')

# Verificar la marca de agua
is_valid = verify_watermark(image, stego_image, model)
print(f"Marca de agua extraída: {extracted_watermark.squeeze().detach().numpy()}")
print(f"Verificación de la marca de agua: {'válida' if is_valid else 'inválida'}")

```
