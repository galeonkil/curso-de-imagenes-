Librerias


```python
import numpy as np
import matplotlib.pyplot as plt
import cv2
import pandas as pd
from google.colab.patches import cv2_imshow
from google.colab import drive
drive.mount('/content/drive')
```

**Carga de Imagenes y template seleccionados**

```python
image=cv2.imread("sample_data/lobo-gris.jpg")
template=cv2.imread("sample_data/template.JPG")
```

**Canbio de color de la iamgen a escala de grises**

```python
image_grey=cv2.cvtColor(image,cv2.COLOR_BGR2GRAY)
template_grey=cv2.cvtColor(template,cv2.COLOR_BGR2GRAY)
```

**Uso del metodo diferencia de cuadrados**

```python
res=cv2.matchTemplate(image_grey,template_grey,cv2.TM_SQDIFF)
```

**Hallar minimo maximo**

```python
min_val,max_val,min_loc,max_loc=cv2.minMaxLoc(res)
```

```python
print(min_val,max_val,min_loc,max_loc)
```

```python
x1,y1=min_loc
x2,y2=max_loc[0]+template.shape[1],max_loc[1]+template.shape[0]
```

**Ploteo de la imagen**

```python
cv2.rectangle(image,(x1,y1),(x2,y2),(0,255,0),3)
cv2_imshow(image)
cv2_imshow(template)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
