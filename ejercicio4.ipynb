```python
import numpy as np
import matplotlib.pyplot as plt
import cv2
import pandas as pd
from google.colab.patches import cv2_imshow
```

Ploteo imagen

```python
imagen=cv2.imread("paisaje-1.jpg")
imagen2=cv2.imread("camara.jpg")
cv2_imshow(imagen)
plt.show()
cv2_imshow(imagen2)
plt.show()
```

Obtencion de pixeles


```python
pixel_values=imagen.reshape((-1,3))

datos=pd.DataFrame(pixel_values,columns=['Red','Green','Blue'])
pixel_values = np.float32(pixel_values)
pixel_red=datos["Red"]
pixel_green=datos["Green"]
pixel_blue=datos["Blue"]
```

Histograma de color

```python
plt.figure(figsize=(10, 5))

plt.subplot(1, 3, 1)
plt.hist(pixel_red, bins=256, color='red', alpha=0.7)
plt.title('Histograma de Rojo')
plt.xlabel('Valor de Rojo')
plt.ylabel('Frecuencia')

plt.subplot(1, 3, 2)
plt.hist(pixel_green, bins=256, color='green', alpha=0.7)
plt.title('Histograma de Verde')
plt.xlabel('Valor de Verde')
plt.ylabel('Frecuencia')

plt.subplot(1, 3, 3)
plt.hist(pixel_blue, bins=256, color='blue', alpha=0.7)
plt.title('Histograma de Azul')
plt.xlabel('Valor de Azul')
plt.ylabel('Frecuencia')

plt.tight_layout()
plt.show()
```

```python
pixel_values2=imagen2.reshape((-1,3))

datos=pd.DataFrame(pixel_values2,columns=['Red','Green','Blue'])
pixel_values2= np.float32(pixel_values)
pixel_red2=datos["Red"]
pixel_green2=datos["Green"]
pixel_blue2=datos["Blue"]
```

```python
imagen_suma = np.zeros_like(imagen, dtype=np.float32)

# Sumar las imágenes pixel por pixel
for i in range(imagen.shape[0]):
    for j in range(imagen.shape[1]):
        for k in range(imagen.shape[2]):
            valor_pixel = float(imagen[i, j, k]) + float(imagen2[i, j, k])

            # Limitar los valores de los píxeles al rango [0, 255]
            if valor_pixel > 255:
                valor_pixel = 255

            # Asignar el valor al píxel correspondiente en imagen_suma
            imagen_suma[i, j, k] = valor_pixel

# Normalizar imagen_suma
imagen_suma_normalizada = (imagen_suma + 1e-7 - np.min(imagen_suma + 1e-7)) / (np.max(imagen_suma + 1e-7) - np.min(imagen_suma + 1e-7)) * 255

# Mostrar la imagen normalizada
cv2_imshow(imagen_suma_normalizada)
plt.show()
```

```python
pixel_values=imagen_suma.reshape((-1,3))

datos=pd.DataFrame(pixel_values,columns=['Red','Green','Blue'])
pixel_values= np.float32(pixel_values)
pixel_red=datos["Red"]
pixel_green=datos["Green"]
pixel_blue=datos["Blue"]
```

```python
plt.figure(figsize=(10, 5))

plt.subplot(1, 3, 1)
plt.hist(pixel_red, bins=256, color='red', alpha=0.7)
plt.title('Histograma de Rojo')
plt.xlabel('Valor de Rojo')
plt.ylabel('Frecuencia')

plt.subplot(1, 3, 2)
plt.hist(pixel_green, bins=256, color='green', alpha=0.7)
plt.title('Histograma de Verde')
plt.xlabel('Valor de Verde')
plt.ylabel('Frecuencia')

plt.subplot(1, 3, 3)
plt.hist(pixel_blue, bins=256, color='blue', alpha=0.7)
plt.title('Histograma de Azul')
plt.xlabel('Valor de Azul')
plt.ylabel('Frecuencia')

plt.tight_layout()
plt.show()
```
