```python
import numpy as np
import matplotlib.pyplot as plt
import cv2
import pandas as pd
from google.colab.patches import cv2_imshow
```

```python
imagen=cv2.imread("flor-morada.jpg")
cv2_imshow(imagen)
plt.show()
```

```python
# Supongamos que 'imagen' es tu matriz de píxeles
valor_a_multiplicar = 3

# Crear una matriz vacía para almacenar los valores de la nueva imagen
nueva_imagen = np.empty_like(imagen)

# Iterar sobre cada elemento de la matriz 'imagen'
for i in range(imagen.shape[0]):
    for j in range(imagen.shape[1]):
        for k in range(imagen.shape[2]):
            # Multiplicar el valor especificado al elemento actual
            nuevo_valor = imagen[i, j, k] * valor_a_multiplicar
            if nuevo_valor > 255:
              nuevo_valor = 255
            # Almacenar el nuevo valor en la matriz de la nueva imagen
            nueva_imagen[i, j, k] = nuevo_valor

# Imprimir la matriz nueva_imagen
cv2_imshow(nueva_imagen)
plt.show()
```

```python
# Supongamos que 'imagen' es tu matriz de píxeles
valor_a_dividir = 3

# Crear una matriz vacía para almacenar los valores de la nueva imagen
nueva_imagen = np.empty_like(imagen)

# Iterar sobre cada elemento de la matriz 'imagen'
for i in range(imagen.shape[0]):
    for j in range(imagen.shape[1]):
        for k in range(imagen.shape[2]):
            # Multiplicar el valor especificado al elemento actual
            nuevo_valor = imagen[i, j, k] / valor_a_dividir
            if nuevo_valor < 0:
              nuevo_valor = 0
            # Almacenar el nuevo valor en la matriz de la nueva imagen
            nueva_imagen[i, j, k] = nuevo_valor

# Imprimir la matriz nueva_imagen
cv2_imshow(nueva_imagen)
plt.show()
```

```python
pixel_values=nueva_imagen.reshape((-1,3))

datos=pd.DataFrame(pixel_values,columns=['Red','Green','Blue'])
pixel_values = np.float32(pixel_values)
pixel_red=datos["Red"]
pixel_green=datos["Green"]
pixel_blue=datos["Blue"]
```

```python
plt.figure(figsize=(10, 5))

plt.subplot(1, 3, 1)
plt.hist(pixel_red, bins=256, color='red', alpha=0.7)
plt.title('Histograma de Rojo')
plt.xlabel('Valor de Rojo')
plt.ylabel('Frecuencia')

plt.subplot(1, 3, 2)
plt.hist(pixel_green, bins=256, color='green', alpha=0.7)
plt.title('Histograma de Verde')
plt.xlabel('Valor de Verde')
plt.ylabel('Frecuencia')

plt.subplot(1, 3, 3)
plt.hist(pixel_blue, bins=256, color='blue', alpha=0.7)
plt.title('Histograma de Azul')
plt.xlabel('Valor de Azul')
plt.ylabel('Frecuencia')

plt.tight_layout()
plt.show()
```
